# Merge

```python
var10=pd.DataFrame({"A":[10,20,30,40],"B":[15,16,17,18]})
var11=pd.DataFrame({"A":[10,20,30,40],"C":[25,26,27,28]})
pd.merge(var10,var11, on="A")
```

```python
var12=pd.DataFrame({"A":[10,20,30,40],"B":[15,16,17,18]})
var13=pd.DataFrame({"A":[10,20,30,50],"C":[25,26,27,28]})
pd.merge(var12,var13, how="right")
```

```python
pd.merge(var12,var13, how="outer", indicator=True)
```

```python
var14=pd.DataFrame({"A":[10,20,30,40],"B":[15,16,17,18]})
var15=pd.DataFrame({"A":[10,20,30,50],"B":[25,26,27,28]})
pd.merge(var14,var15, left_index=True, right_index=True, suffixes=("name","Python"))
```

# Concat

```python
sr1=pd.Series([1,2,3,4,5])
sr2=pd.Series([11,12,13,14,15])
pd.concat([sr1,sr2])
```

```python
dd=pd.DataFrame({"A":[10,20,30,40],"B":[15,16,17,18]})
de=pd.DataFrame({"A":[10,20,30,50],"B":[25,26,27,28]})
pd.concat([dd,de])
```

```python
dd=pd.DataFrame({"A":[10,20,30,40],"B":[15,16,17,18]})
de=pd.DataFrame({"A":[10,20,30,50],"c":[25,26,27,28]})
pd.concat([dd,de],axis=1)
```

```python
dd=pd.DataFrame({"A":[10,20,30,40],"B":[15,16,17,18]})
de=pd.DataFrame({"A":[10,20],"c":[25,26]})
pd.concat([dd,de],axis=1, join="inner",)
```

```python
dd=pd.DataFrame({"A":[10,20,30,40],"B":[15,16,17,18]})
de=pd.DataFrame({"A":[10,20,30,50],"c":[25,26,27,28]})
pd.concat([dd,de],axis=1,keys=["dd","de"])
```

# Join

```python
d1=pd.DataFrame({"A":[10,20,30,40],"B":[15,16,17,18]})
d2=pd.DataFrame({"C":[101,201,301,501],"D":[25,26,27,28]})
d1.join(d2)
```

```python
d1=pd.DataFrame({"A":[10,20,30,40],"B":[15,16,17,18]},index=["a","b","c","d"])
d2=pd.DataFrame({"C":[101,201],"D":[25,26]},index=["a","b"])
d1.join(d2)
```

```python
d1=pd.DataFrame({"A":[10,20,30,40],"B":[15,16,17,18]},index=["a","b","c","d"])
d2=pd.DataFrame({"C":[101,201],"D":[25,26]},index=["a","b"])
d2.join(d1,how="inner")
```

```python
d3=pd.DataFrame({"A":[10,20,30,40],"B":[15,16,17,18]},index=["a","b","c","d"])
d4=pd.DataFrame({"C":[101,201],"B":[25,26]},index=["a","b"])
d4.join(d3,how="outer",lsuffix="C", rsuffix="C")
```

# Pivot()

```python
p=pd.DataFrame({"Days":[1,2,3,4,5,6],
                "St_name":["a","b","a","a","b","b"],"Eng":[10,12,14,15,13,19],"Maths":[17,19,20,10,12,18]})
p
```

```python
p.pivot(index="Days",columns="St_name",values="Eng")
```

```python
p1=pd.DataFrame({"Days":[1,1,1,1,2,2],
                "St_name":["a","b","a","a","b","b"],"Eng":[10,12,14,15,13,19],"Maths":[17,19,20,10,12,18]})
p1
```

```python
p1.pivot_table(index="St_name",columns="Days",
              aggfunc="mean", margins="True")
```
