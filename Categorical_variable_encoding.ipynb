#### Ordinal variables
It is a categorical variable in which the categories can be meaningfully ordered.
When the ordinal variable convert into the number then it has mathematical value.

#### Nominal Variable
<p> Nominal variable is same like ordinal variable but there is nothing that indicates an intrinsic order of the labels and in principle</p>
<p>All labels are equal no order</p>
<p>No Mathematical Value</p>

### Lable Encoding vs Ordinal Encoding
<p><b>Label Encoding</b> apply on ordinal and nominal categorical variables</p>
<p><b>Ordinal Encoding</b> apply on ordinal categorical variables</p>
