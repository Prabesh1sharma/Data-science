# Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

# Lineplot

```python
csv= sns.load_dataset("penguins") #can use .head() for small data set as you need
csv
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=csv)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=csv,hue="sex")
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=csv,hue="sex",style="sex",palette="Accent",markers=["o","<"], 
            dashes=False,legend="full")
plt.grid()
plt.title("Penguins")
plt.show()# we can use size=60 parameter for size
```

# Barplot

```python
sns.barplot(x="island",y="flipper_length_mm",data=csv)
plt.show()
```

```python
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="flipper_length_mm",data=csv,hue="sex",order=order_1,)
plt.show()
```

```python
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="flipper_length_mm",data=csv,hue="sex",order=order_1,
           hue_order=["Female","Male"],errorbar=('ci', 100),)
plt.show()
```

```python
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="bill_length_mm",y="flipper_length_mm",data=csv,orient="v",hue="sex",palette="icefire")
plt.show()
```

```python
#sns.set(style="darkgrid")
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="flipper_length_mm",data=csv,orient="v",hue="sex",
           saturation=10,err_kws={'color': 'b'},capsize=0.1,dodge=True,alpha=0.5)
plt.show()
```

# Count plot

<p>->Countplot plots the count of the number of records by category.</p>
<p>->Barplot plots a value or metric for each category (by default, barplot plots the mean of variable).</p>

```python
csv2=sns.load_dataset("tips")
csv2
```

```python
sns.countplot(x="sex",data=csv2,hue="smoker",palette="rainbow_r",saturation=0.9)
plt.show()
```

# Violin Plot

```python
sns.violinplot(x="day",y="total_bill",data=csv2,hue="time")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=csv2,hue="sex",linewidth=2,palette="cool",saturation=0.9,
              )
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data=csv2,order=["Dinner","Lunch"],inner="quart")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=csv2,hue="sex",split=True)
plt.show()
```

```python
sns.violinplot(x=csv2["total_bill"])
plt.show()
```

# Pair plot

```python
sns.pairplot(csv2,hue="sex")
plt.show()
```

```python
sns.pairplot(csv2,vars=["total_bill","tip"],hue="sex",palette="hsv_r")
plt.show()
```

```python
sns.pairplot(csv2,hue="sex",x_vars=["total_bill","tip"])
plt.show()
```

```python
sns.pairplot(csv2,hue="sex",kind="reg")
plt.show()
```

# Strip plot

# Box Plot

```python
sns.set(style="whitegrid")

sns.boxplot(x="day",y="total_bill",data=csv2,hue="time",palette='dark:w',showmeans=True,meanprops={"marker":"+","markeredgecolor":"b"})
plt.show()
```

```python
sns.set(style="whitegrid")

sns.boxplot(y=csv2["total_bill"])
plt.show()
```

# Factor plot replace by Catplot

```python
sns.catplot(x="day",y="tip",data=csv2,hue="sex",kind="boxen") 

plt.show()
```

# Styling plot

<b><p>Seaborn Figure style</p>
<p>Removing Axes Spines</p>
<p>Scale and Context</p></b>

```python
sns.set_style("darkgrid")
sns.barplot(x="day",y="total_bill",data=csv2)
plt.show()
```

```python
sns.set_style("white")
plt.figure(figsize=(12,5))
sns.barplot(x="day",y="total_bill",data=csv2)
sns.despine()

plt.show()
```

```python
sns.set_style("white")
sns.set_context("talk",font_scale=1)
sns.barplot(x="day",y="total_bill",data=csv2)
plt.show()
```

# Multiple plot(Facet-Grid)

```python
fg=sns.FacetGrid(csv2,col="day",hue="sex")

fg.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()
```
