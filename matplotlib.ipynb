# Bar plot

```python
x=["python","c","c++","java"]
y=[85,70,60,82]
plt.xlabel("Langauage",fontsize=15)
plt.ylabel("Number")
plt.title("DEMO")
plt.bar(x,y)
plt.show()
```

```python
x=["python","c","c++","java"]
y=[85,70,60,82]
plt.xlabel("Langauage",fontsize=15)
plt.ylabel("Number")
plt.title("DEMO")
c=["y","b","g","m"]
plt.bar(x,y,width=0.4,color=c,align="center",edgecolor="r",linewidth=2,linestyle="-",alpha=0.4,label=["Python","c","c++","Java"])
plt.legend()
plt.show()
```

```python
x=["python","c","c++","java"]
y=[85,70,60,82]
z=[20,30,40,60]
plt.xlabel("Langauage",fontsize=15)
plt.ylabel("Number")
plt.title("DEMO")
c=["y","b","g","m"]
plt.bar(x,y,width=0.4,color="r",alpha=1,label="Popularity")
plt.bar(x,z,width=0.4,color="y",alpha=1,label="Popularity1")
plt.legend()
plt.show()
```

```python
import numpy as np

x=["python","c","c++","java"]
y=[85,70,60,82]
z=[20,30,40,60]
p=np.arange(len(x))
width=0.2
p1=[j+width for j in p]

plt.xlabel("Langauage",fontsize=15)
plt.ylabel("Number")
plt.title("DEMO")

plt.bar(p,y,width,color="r",alpha=1,label="Popularity")
plt.bar(p1,z,width,color="y",alpha=1,label="Popularity1")

plt.xticks(p+width/2,x,rotation=0)
plt.legend()
plt.show()
```

```python
x=["python","c","c++","java"]
y=[85,70,60,82]
z=[20,30,40,60]
plt.xlabel("Langauage",fontsize=15)
plt.ylabel("Number")
plt.title("DEMO")

plt.barh(x,y,color="r",alpha=1,label="Popularity")
plt.barh(x,z,color="y",alpha=1,label="Popularity1")

plt.legend()
plt.show()
```

# Pie Plot

```python
x=[10,20,30,40]
y=["c","c++","java","python"]
plt.pie(x,labels=y)
plt.show()
```

```python
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.4,0.0,0.0,0.0]
c=["r","g","y","b"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=1,labeldistance=1.1,startangle=0,
       textprops={"fontsize":15},counterclock=True,wedgeprops={"linewidth":5,"edgecolor":"w"},
       rotatelabels=True)
plt.title("Piechart")
plt.legend(loc=2)
plt.show()
```

# Stack Plot /Area plot

```python
x=[10,20,30,40,50,60,70]
area1=[10,20,40,70,60,90,30]
area2=[30,60,78,90,50,10,44]
area3=[14,88,40,80,40,40,34]
l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l,colors=["r","y","b"])
plt.legend(loc=2)
plt.show()
```

```python
x=[10,20,30,40,50,60,70]
area1=[10,20,40,70,60,90,30]
area2=[30,60,78,90,50,10,44]
area3=[14,88,40,80,40,40,34]
l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l,colors=["r","y","b"],baseline="zero")
plt.title("Demo")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend(loc=2)
plt.show()
```

# Fill_Between Plot

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]
plt.plot(x,y,color="r")

plt.fill_between(x=[2,4],y1=2,y2=4)

plt.title("Fill Between")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])
plt.plot(x,y,color="r")

plt.fill_between(x,y,color="g",where=(x>=2)&(x<=4),alpha=0.5)

plt.title("Fill Between")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

# Savefig

```python
x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])
plt.subplot(2,2,1)
plt.plot(x,y,color="r")
plt.subplot(2,2,2)
plt.pie([1],colors="b")

x1=[10,20,30,40]
plt.subplot(2,2,3)
plt.pie(x)
x2=["python","c","c++","java"]
y2=[85,70,60,82]
plt.subplot(2,2,4)
plt.bar(x2,y2)
plt.savefig("demo1",dpi=2000)
plt.show()
```

```python
x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])
plt.plot(x,y,color="r")
plt.savefig("demo2",dpi=2000,facecolor="b",transparent=False)
plt.show()
```
